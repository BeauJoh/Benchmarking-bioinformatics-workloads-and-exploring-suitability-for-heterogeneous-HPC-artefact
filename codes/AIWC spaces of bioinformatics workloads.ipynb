Load the data:

```python
%cd /workspace/codes/

import glob
import pandas as pd

list_of_csv = glob.glob("/OpenDwarfs/build/aiwc_needle_opencl_shared_1_*.csv")

frames = []
i = 0
for csv_file in list_of_csv:
    x = pd.read_csv(csv_file, sep=",")
    x['application'] = 'nw'
    x['kernel'] = 'shared_1'
    x['invocation'] = i
    x['size'] = 'small'
    frames.append(x)
    i = i + 1
    
metrics = pd.concat(frames)

list_of_csv = glob.glob("/OpenDwarfs/build/aiwc_needle_opencl_shared_2_*.csv")

i = 0
for csv_file in list_of_csv:
    x = pd.read_csv(csv_file, sep=",")
    x['application'] = 'nw'
    x['kernel'] = 'shared_2'
    x['invocation'] = i
    x['size'] = 'large'
    frames.append(x)
    i = i + 1
    
metrics = pd.concat(frames)

#TODO: Add number of kernel invocations to kiviat diagram -- as a compute metric?
```

shared_1 kernel -- Studentize/Normalize metrics:

```python
%%R -i metrics -w 800 -h 800 -u px

library('fmsb')
library('cowplot')
source('./kiviat-plotter.R')
source('./reordering-codes.R')

#Normalize according to previously collected metrics
load('/opencl-predictions-with-aiwc/data/intermediate/featdata.Rda')

featdata.all <- rbind(featdata.all,metrics)

#Reorder the dataframes
x <- reorder_and_subset(featdata=featdata.all,size='small',kernel="shared_1")
x <- data.frame(t(colMeans(x)))
y <- x
nv <- subset(reorder_features(featdata.all),select=-c(size,kernel,application,invocation))

plot_kiviat(y,nv,colour="black",colour_spokes=TRUE)
```
