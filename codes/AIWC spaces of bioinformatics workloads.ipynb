## Dynamic Programming

### Needleman-Wunsch -- nw

Generate nw results:

```python
%cd /OpenDwarfs/build
! $OCLGRIND_BIN --workload-characterisation ./needle -p 0 -d 0 -t 0 -- 2048 10
! rm -f lsb.*.r0* #we aren't interested in runtime data
```

Load the data:

```python
%cd /workspace/codes/

import glob
import pandas as pd

list_of_csv = glob.glob("/OpenDwarfs/build/aiwc_needle_opencl_shared_1_*.csv")

frames = []
i = 0
for csv_file in list_of_csv:
    x = pd.read_csv(csv_file, sep=",")
    x['application'] = 'nw'
    x['kernel'] = 'shared_1'
    x['invocation'] = i
    x['size'] = 'small'
    frames.append(x)
    i = i + 1
    
metrics = pd.concat(frames)

list_of_csv = glob.glob("/OpenDwarfs/build/aiwc_needle_opencl_shared_2_*.csv")

i = 0
for csv_file in list_of_csv:
    x = pd.read_csv(csv_file, sep=",")
    x['application'] = 'nw'
    x['kernel'] = 'shared_2'
    x['invocation'] = i
    x['size'] = 'small'
    frames.append(x)
    i = i + 1
    
metrics = pd.concat(frames)

#TODO: Add number of kernel invocations to kiviat diagram -- as a compute metric?
```

shared_1 and shared_2 kernel -- Studentize/Normalize metrics:

```python
%%R -i metrics -w 800 -h 800 -u px

library('fmsb')
library('cowplot')
source('./kiviat-plotter.R')
source('./reordering-codes.R')

#Reorder the dataframes
load('/opencl-predictions-with-aiwc/data/intermediate/featdata.Rda')
new.featdata <- rbind(featdata.all,metrics)
nv <- subset(reorder_features(new.featdata),select=-c(size,kernel,application,invocation))

x <- reorder_and_subset(featdata=new.featdata,size='small',kernel="shared_1")
x <- data.frame(t(colMeans(x)))
y <- x

x <- reorder_and_subset(featdata=new.featdata,size='small',kernel="shared_2")
x <- data.frame(t(colMeans(x)))
y <- rbind(y,x)

k_names <- c("shared_1","shared_2")

#Normalize according to previously collected metrics
plot_kiviat(y,nv,colour="black",colour_spokes=TRUE,all_kernels=TRUE,kernel_names=k_names)
```

Note: shared_1 and shared_2 kernels have near identical feature-spaces:

```python
%%R -i metrics -w 800 -h 800 -u px
print(y)
```

This is due to the structure of these kernels -- thus also the characteristics -- being similar. As shown below

```python
%cd /OpenDwarfs/build
!sed -n '21,72p' ./needle_kernel.cl > shared_1.txt
!sed -n '75,123p' ./needle_kernel.cl > shared_2.txt
!diff shared_1.txt shared_2.txt
!rm shared_1.txt shared_2.txt
%cd /workspace/codes/
```

Only the shared_2 kernel is supported on the lower right part of the matrix, and supports the us of work-group (in the form of block_width) which means a different indexing is used.
